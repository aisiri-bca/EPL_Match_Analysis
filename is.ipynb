```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt

# --- Title ---
st.title("Premier League Matches Data")
st.write("Explore match results, goals, and team statistics from past seasons.")

# --- Load CSV ---
df = pd.read_csv("premier-league-matches.csv")

# --- Sidebar Filter ---
season = st.sidebar.selectbox("Select Season", sorted(df['Season_End_Year'].unique()))
filtered_df = df[df['Season_End_Year'] == season]

st.subheader(f"Matches in Season {season}")
st.dataframe(filtered_df)

# --- Summary Stats ---
st.subheader("Summary Statistics")
st.write("Goals statistics in this season:")
st.write(filtered_df[['HomeGoals', 'AwayGoals']].describe())

st.write("Full-time result counts (H=Home Win, D=Draw, A=Away Win):")
st.bar_chart(filtered_df['FTR'].value_counts())

# --- Top Home Teams by Goals ---
st.subheader("Top 10 Home Teams by Goals Scored")
top_home_teams = filtered_df.groupby('Home')['HomeGoals'].sum().sort_values(ascending=False).head(10)
st.bar_chart(top_home_teams)

# --- Matches per Week Plot ---
st.subheader("Matches per Week")
matches_per_week = filtered_df.groupby('Wk').size()
st.line_chart(matches_per_week)

# --- Optional: Show Raw Data ---
if st.checkbox("Show Raw Data"):
    st.dataframe(df)
```
